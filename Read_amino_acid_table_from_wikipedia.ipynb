```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def table_to_pandas(table,header_count):
 
    headers = [header.text for header in table.find_all('th')]
    header=headers[:header_count]
    indices=headers[header_count:]
    
    rows = []
    for aa,row in zip(indices,table.find_all('tr')):
         rows.append([aa]+[val.text for val in row.find_all('td')])
    return pd.DataFrame(rows,columns=header)

pre_soup=requests.get("https://en.wikipedia.org/wiki/Proteinogenic_amino_acid")
soup=BeautifulSoup(pre_soup.content,"lxml")
tables=soup.findAll("table")
```

```python
chemical_properties_table=table_to_pandas(tables[0],7) #table 0 is the chemical data
chemical_properties_table=chemical_properties_table.dropna()
chemical_properties_table.head()

side_chain_table=table_to_pandas(tables[1],12) #table 1 is the side chain data
side_chain_table=side_chain_table.dropna()
aa_data=side_chain_table.merge(chemical_properties_table,on=["Short","Amino acid","Abbrev."])
aa_data.head()
```

```python
aa_standard=aa_data[(aa_data["Short"]!="O") & (aa_data["Short"]!="U")]
aa_standard.head()
```

```python
pd.DataFrame.to_csv(aa_data,"Amino_acid_properties_wiki_data.csv")
```
